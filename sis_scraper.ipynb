```python
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
import time
```

```python
driver = webdriver.Chrome(ChromeDriverManager().install()) 
```

```python
url = 'https://sisdatahub.com/leaderboards/QB' # works for /WR and /CB as well, should work for all positions in theory
```

```python
driver.get(url)
time.sleep(10) # wait a little bit to make sure page fully loads.
```

```python
# should be able to switch to other 'sub-tables' in chromedriver browser - 'adv passing' for qbs, 'pass blocking' for OLs, etc -
# and pull those by running this + below cells

cells = driver.find_elements_by_class_name('tabulator-cell')
```

```python
players = {}
temp_name = ''
temp_fields = {}
first_row_done = False
for cell in cells:
    if cell.get_attribute('tabulator-field') == 'playerName' or cell.get_attribute('tabulator-field') == 'player':
        if first_row_done:
            players[temp_name] = temp_fields
        else:
            first_row_done = True
        temp_name = cell.text
        temp_fields = {}
    else:
        attribute = cell.get_attribute('tabulator-field')
        value = cell.text
        if ',' in value:
            value = value.replace(',', '')
        try:
            value = float(value)
        except ValueError:
            pass
        temp_fields[attribute] = value
```

```python
df = pd.DataFrame(players).transpose()
```
